# pandas 데이터 가공

```python
import pandas as pd
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"]},
                  index=[0,1,2,3])

df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                   "B":["B4","B5","B6","B7"],
                   "C":["C4","C5","C6","C7"],
                   "D":["D4","D5","D6","D7"]},
                  index=[4,5,6,7])
```

```python
df1
```

```python
df2
```

## 1) 아래로 합치기 : concat / append

### concat

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

### append

```python
df1.append(df2)
```

### 컬럼 이름이 다를 경우
컬럼 이름 매칭 작업은 꼭 필요하다

```python
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"]},
                  index=[0,1,2,3])

df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                   "B":["B4","B5","B6","B7"],
                   "CC":["C4","C5","C6","C7"],
                   "DD":["D4","D5","D6","D7"]},
                  index=[4,5,6,7])
```

```python
pd.concat([df1,df2])
```

```python
df1.append(df2, sort=False)
```

### 컬럼이 다르고 인덱스도 일부 겹치는 경우

```python
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"]},
                  index=[0,1,2,3])

df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                   "B":["B4","B5","B6","B7"],
                   "CC":["C4","C5","C6","C7"],
                   "DD":["D4","D5","D6","D7"]},
                  index=[2,3,6,7])
```

```python
pd.concat([df1,df2])
```

```python
# index의 정리
pd.concat([df1,df2], ignore_index=True)
```

## 2) 옆으로 불이기 : axis = 1

```python
import pandas as pd
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"]},
                  index=[0,1,2,3])

df2 = pd.DataFrame({"AA":["A4","A5","A6","A7"],
                   "BB":["B4","B5","B6","B7"],
                   "CC":["C4","C5","C6","C7"],
                   "DD":["D4","D5","D6","D7"]},
                  index=[0,1,2,3])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
# append에서 axis는 없음
df1.append(df2, axis=1)
```

### 컬럼과 인덱스가 일부 겹칠때
컬럼 이름은 중복이 있어도 무시하고 옆으로 붙이지만<br>
인덱스는 중복을 인식하고, 중복이 없으면 추가한다

```python
import pandas as pd
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"]},
                  index=[0,1,2,3])

df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                   "B":["B4","B5","B6","B7"],
                   "CC":["C4","C5","C6","C7"],
                   "DD":["D4","D5","D6","D7"]},
                  index=[2,3,5,7])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1).dropna()
```

##### 원하는 컬럼만 선택한다 해도, pandas는 기본적으로 중복된 것을 모두 다 출력한다.

```python
pd.concat([df1,df2],axis=1).loc[:,["A","B","CC"]]
```

##### 특정 인덱스를 설정하는 법 : .reindex(...)

```python
pd.concat([df1,df2],axis=1).reindex(df2.index)
```
