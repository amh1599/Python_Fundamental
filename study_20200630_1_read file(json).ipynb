### DataFrame 만들기

```python
# 모든 정보를 담을 DF
movieIdDF = pd.DataFrame()

# 1갸의 영화 정보들에 대해서 처리할 DF
movieIdDF_temp = pd.DataFrame()
movieIdDF_temp = movieIdDF_temp.append({
    "movieCd":"",
    "movieNm":"",
    "movieNmEn":"",
    "openDt":"",
    "directorpeopleNm":""},
    ignore_index=True)

# 받아온 데이터의 수
num = len(movieIdData["movieListResult"]["movieList"])

# movieIdDF_temp.columns.get_loc("movieCd")  =>  1

temp_movcd_idx = movieIdDF_temp.columns.get_loc("movieCd")
temp_movnm_idx = movieIdDF_temp.columns.get_loc("movieNm")
temp_movnmen_idx = movieIdDF_temp.columns.get_loc("movieNmEn")
temp_open_idx = movieIdDF_temp.columns.get_loc("openDt")
temp_dir_idx = movieIdDF_temp.columns.get_loc("directorpeopleNm")

index=movieIdData["movieListResult"]["movieList"]
for i in range(0,num):
    # 1. movieIdDF_temp에 i번째 영화 정보 담기
    movieIdDF_temp.iat[0,temp_movcd_idx]=index[i]["movieCd"]
    movieIdDF_temp.iat[0,temp_movnm_idx]=index[i]["movieNm"]
    movieIdDF_temp.iat[0,temp_movnmen_idx]=index[i]["movieNmEn"]
    movieIdDF_temp.iat[0,temp_open_idx]=index[i]["openDt"]
    if index[i]["directors"]:
        movieIdDF_temp.iat[0,temp_dir_idx]=index[i]["directors"][0]["peopleNm"]
    else:
        movieIdDF_temp.iat[0,temp_dir_idx]=""
    
    # 2. movieIdDF_temp에 있는 영화정보를 movieIdDF에 추가
    movieIdDF = movieIdDF.append(movieIdDF_temp, ignore_index=True)
                                                                               
    # 3. movieIdDF_temp의 영화정보 초기화
    movieIdDF_temp.iloc[0,:]=""
                                                                               
movieIdDF
```

##### year, month, day의 추가

```python
movieIdDF["year"]=movieIdDF["openDt"].apply(lambda x : str(x)[0:4])
movieIdDF["month"]=movieIdDF["openDt"].apply(lambda x : str(x)[4:6])
movieIdDF["day"]=movieIdDF["openDt"].apply(lambda x : str(x)[6:8])
movieIdDF
```

##### openDt 기준 정렬

```python
movieIdDF = movieIdDF.sort_values(by="openDt")
movieIdDF
```

##### movieCd 가 고유 번호이므로 index로 설정

```python
movieIdDF= movieIdDF.set_index("movieCd")
movieIdDF
```

##### column의 순서 설정

```python
col_list=["openDt","year","month","day","movieNm","movieNmEn","directorpeopleNm"]
movieIdDF = movieIdDF.loc[:, col_list]
movieIdDF
```

##### excel로 출력

```python
# 한글이 깨지므로 encoding="ms949" or "cp949" 해줌
movieIdDF.to_csv("kobis20200630.csv", encoding="ms949")
```
