## 데이터 합치기 : join

```python
import pandas as pd
import numpy as np
```

```python
baseball_player = pd.DataFrame(
    {
        "player_name":
        ["권혁","이범호","필","박병호","양준혁"],
        "team_No":
        [1993,2001,2001,2010,None]
    })
baseball_team = pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
        "team_id":[1993,1999,2001,2010]
    })
```

```python
baseball_player
```

```python
baseball_team
```

### merge(left, right, how="inner", on=None, left_on=None, right_on=None, left_index=False, right_indes=False, sort=True, suffixes=('_x','_y'), copy=True, indicator=False)

### 1) inner (둘 다 공통 된 정보)

```python
pd.merge(baseball_player, baseball_team,
         how="inner",
         left_on="team_No", 
         right_on="team_id")
```

### 2) left (왼쪽과 공통 된 정보)

```python
pd.merge(baseball_player, baseball_team,
         how="left",
         left_on="team_No", 
         right_on="team_id")
```

### 3) right (오른쪽과 공통 된 정보)

```python
pd.merge(baseball_player, baseball_team,
        how="right",
        left_on="team_No",
        right_on="team_id")
```

### 4) left excluding join (왼쪽에만 있는 정보)

```python
temp =pd.merge(baseball_player, baseball_team,
        how="left",
        left_on="team_No",
        right_on="team_id")
temp
```

```python
temp[temp["team_id"].isnull()]
```

### 5) right excluding join (오른쪽에만 있는 정보)

```python
temp =pd.merge(baseball_player, baseball_team,
        how="right",
        left_on="team_No",
        right_on="team_id")
temp
```

```python
temp[temp["team_No"].isnull()]
```

### 6) outer (양쪽에 있는 모든 정보)

```python
pd.merge(baseball_player, baseball_team,
        how="outer",
        left_on="team_No",
        right_on="team_id")
```

### 7) outer excluding join (공통 된 정보를 제외한 모든 정보)

```python
temp=pd.merge(baseball_player, baseball_team,
        how="outer",
        left_on="team_No",
        right_on="team_id")
temp
```

```python
temp[temp["team_No"].isnull() | temp["team_id"].isnull()]
```

## 인덱스와 데이터의 join : left_index, right_index

```python
baseball_player=pd.DataFrame({
    "palyer_Nmae":["권혁","이범호","필","박병호","양준혁"],
    "team_No":[1993,2001,2001,2010,None]},
    index=["p_01","p_02","p_03","p_04","p_05"])

baseball_player.index.name="player_id"

baseball_team = pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"]},
    index=[1993,1999,2001,2010])
```

```python
baseball_player
```

```python
baseball_team
```

```python
#inner join
pd.merge(baseball_player, baseball_team,
        how="inner",
        left_on="team_No",
        right_index=True)
```

```python
# 1) left join --> 선수들 정보 다 있는 상태에서 팀 정보 가져다 붙이기
# 2) right join --> 팀 정보들을 기준으로 선수들에 대한 정보 가져다 붙이기
```

```python
pd.merge(baseball_player, baseball_team,
        how="left",
        left_on="team_No",
        right_index=True)
```

```python
pd.merge(baseball_player, baseball_team,
        how="right",
        left_on="team_No",
        right_index=True)
```

```python
temp = pd.merge(baseball_player, baseball_team,
        how="outer",
        left_on="team_No",
        right_index=True)
temp
```

```python
temp[temp.index.isnull() | temp["team_No"].isnull()]
```
