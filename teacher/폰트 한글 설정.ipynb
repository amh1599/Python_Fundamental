```python
%matplotlib inline
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
from IPython.display import Image
import urllib.request
```

```python
movieIdURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=430156241533f1d058c603178cc3ca0e"
res = urllib.request.urlopen(movieIdURL)
api_movie_data = json.loads(res.read().decode("utf-8"),encoding="utf-8")
```

```python
json_movie_temp = pd.DataFrame()
json_movie_temp = json_movie_temp.append({"구분":"","년":"","월":"","일":"","요일":"","지역-시도":"","성별":"","연령대":"","이용_비율(%)":""}, ignore_index=True)
json_movie_tot = pd.DataFrame()
for i in range(0,len(api_movie_data["movieListResult"]["movieList"])):
    
    json_movie_temp.iat[0,json_movie_temp.columns.get_loc('구분')] = api_movie_data["movieListResult"]["movieList"][i]['typeNm']
    json_movie_temp.iat[0,json_movie_temp.columns.get_loc('년')] = api_movie_data["movieListResult"]["movieList"][i]['openDt']
      #concate로 수정할거!!!!!!
    json_movie_tot = json_movie_tot.append(json_movie_temp,ignore_index=True)   
    for j in range(0,len(json_movie_temp.columns)):
        json_movie_temp.iat[0,j] = ""
json_movie_tot.head()
```

```python
ax = plt.figure(figsize=(10,5))
ax=sns.countplot(x = '구분',data = json_movie_tot)
plt.show()
```
