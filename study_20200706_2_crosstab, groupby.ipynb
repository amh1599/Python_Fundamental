```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("sampledata/11_sales-funnel.xlsx")

data["Status"]=data["Status"].astype("category")
data["Status"].cat.set_categories(
    ["won","pending","presented","declined"],
    inplace=True)
```

# crosstab
.crosstab(가로줄, 세로줄)  -> 리스트 사용 가능

```python
pd.crosstab(data["Name"], data["Price"])
```

```python
# 가로 -> Manager, Rep
# 세로 -> Price, Quantity
pd.crosstab([data["Manager"],data["Rep"]],
            [data["Price"],data["Quantity"]],
            margins=True)
```

```python
pd.crosstab([data["Manager"],data["Rep"]], data["Product"],
            values=data["Price"],
            aggfunc=[len,np.mean])
```

### pivot table과는 다른 방법으로 NaN을 처리한다

```python
pd.crosstab([data["Manager"],data["Rep"]], data["Product"],
            values=data["Price"],
            aggfunc=[len,np.mean]).fillna(0)
```

### 데이터를 월별로 묶기 (groupby)

```python
# 기존 사용하던 방법
for i in data["month"].unique():
    data[data["month"]==i]
```

```python
# groupby 사용
data.groupby(["month"])
```

```python
data.groupby(["month"]).groups.keys()
```

```python
data.groupby(["month"]).groups.values()
```

```python
data.groupby(["month"]).groups["2014-11"]
```

```python
for name, group in data.groupby("month"):
    print(name)
    print(group)
    print()
```

```python
# 각 key값을 기준으로 5개씩 출력
data.groupby("month").head()
```

```python
# 첫번째 key값의 5개만 출력
data.groupby("month").first()
```

```python
data.groupby("month").mean()
```

```python
# 위 결과 중 duration만 출력
data.groupby("month")["duration"].mean()
```

```python
data.groupby("month")["duration"].sum()
```

#### Q) item의 값이 call인 데이터중에서 통신사별로(network) duration의 값의 총 합

```python
data[data["item"]=="call"].groupby("network")["duration"].sum()
```

```python
data.groupby(["month","item"])["date"].count()
```

#### 기준 : month-item
#### 보고자 하는 값 : duration,network_type, date
#### duration -> 합,   network_type -> 건수,   date -> 제일 처음 값

```python
data.groupby(["month","item"]).agg({"duration" : "sum",
                                    "network_type":"count",
                                    "date":"first"})
```

### pivot table 버전

```python
pd.pivot_table(data,
               index=["month","item"],
               aggfunc={"duration":"sum",
                        "network_type":"count",
                        "date":"first"})
```
